```python
# define the menu of cafe
menu = {
    "Pizza": 60,
    "Pasta": 40,
    "Fried rice": 50,
    "Coffee": 50,
    "Burger": 50
}

# Greet
print("Welcome to PYTHON RESTAURANT")
print("Pizza: RS60\nPasta: RS40\nFried rice: RS50\nCoffee: RS50\nBurger: RS50")

order_total = 0

# Get the first item and make it case-insensitive
item_1 = input("Enter the name of the item you want to order: ").strip().lower()

# Check if the item exists in the menu (case-insensitive)
menu_items_lower = {key.lower(): key for key in menu}  # Create a lowercase mapping of menu items

if item_1 in menu_items_lower:
    actual_item = menu_items_lower[item_1]  # Get the correct case version of the item
    order_total += menu[actual_item]  # Add item price to order total
    print(f"Your item {actual_item} has been added to your order.")
else:
    print(f"Ordered item {item_1} is not available!")

# Ask for another item
another_order = input("Do you want to add another item? (Yes/No): ").strip().lower()

if another_order == "yes":
    item_2 = input("Enter the name of the second item: ").strip().lower()

    if item_2 in menu_items_lower:
        actual_item = menu_items_lower[item_2]  # Get the correct case version of the item
        order_total += menu[actual_item]  # Add item price to order total
        print(f"Item {actual_item} has been added to your order.")
    else:
        print(f"Ordered item {item_2} is not available!")

# Print the total order cost
print(f"Your total order cost is: RS{order_total}")
```
